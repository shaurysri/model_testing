```python
from transformers import pipeline
import pandas as pd
import numpy as np

class racism_detector():
    def __init__(self):
        self.model= pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli")
        self.class_labels=["Racism", "No Racism"]
        
        self.hobby_labels=['travelling', 'camping','music','reading','writing','art/drawing','fishing',
                            'collecting','Journaling','cooking','gardening','hiking', 'dancing','photography',
                            'singing','dancing','painting','yoga','sports','swimming','horse riding','skating',
                            'acting','baking','origami','coding','designing','programming','gaming','playing','cycling',
                            'jogging','woodworking','calligraphy','knitting','fishing','birdwatching','shopping',
                            'rafting','astronomy','star gazing','pottery','organic farming','business','hacking','makeup','None']
        
    def predict(self,text):
        result=self.model(text,self.class_labels)
        return {"label":result['labels'][0],"score":result['scores'][0]}
        #print("Label:", result['labels'][0],"Score: ",result['scores'][0])
    def predict_hobby(self,text):
        result=self.model(text,self.hobby_labels)
        print("LAbel:", result['labels'][0:5],"Score: ",result['scores'][0:5])
```

```python
df = pd.read_csv(datasetpath)
df.head()
```

```python
racism = df[df['cyberbullying']==1][['text','cyberbullying']]
```

```python
racism.head()
```

```python
model = racism_detector()
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
pred = []
for text in racism['text'].values:
    test = model.predict(text)
    pred.append(test)
```

```python
Pred = []
for t in pred:
  if(t['label']=='Racism'):
    Pred.append(1)
  else:
    Pred.append(0)
```

```python
print(classification_report(racism['cyberbullying'].values,Pred))
```

```python
print(confusion_matrix(racism['cyberbullying'].values,Pred))
```
