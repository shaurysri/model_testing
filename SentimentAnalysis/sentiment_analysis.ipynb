```python
class sentimentAnalyser():
    def __init__(self):
        self.model=pipeline('sentiment-analysis')

    def predict(self,text):
        result=self.model(text)
        return result
```

```python
#target: the polarity of the tweet (0 = negative, 2 = neutral, 4 = positive)
```

```python
header = ['target','ids','date','flag','user','text']
```

```python
df = pd.read_csv('../input/sentiment140/training.1600000.processed.noemoticon.csv',encoding='ISO-8859-1',names=header)
```

```python
df.head()
```

```python
df = df[['text','target']]
```

```python
df['target'] = df['target'].map({0:'negtaive',2:'neutral',4:'postive'})
```

```python
df['target'].value_counts()
```

## Since the dataset is too large hence testing on random 50000 samples

```python
DF = df.sample(n = 50000,random_state = 101)
```

```python
DF.reset_index(drop=True,inplace=True)
DF['target'].value_counts()
```

```python
model = sentimentAnalyser()
```

```python
DF['target'] = DF['target'].map({"negtaive":'negative',"postive":"positive"})
```

```python
DF[['text','target']]
```

```python
pred = []
for text in DF['text'].values:
    test = model.predict(text)
    pred.append(test[0]['label'].lower())
```

```python
print(classification_report(DF['target'].values,pred))
```

```python
print(confusion_matrix(DF['target'].values,pred))
```
